```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

plt.plot(x,y)
plt.title("My Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1, label="sin(x)", color="blue", linestyle="--", linewidth=2,marker="o")
plt.plot(x,y2, label="cos(x)", color="red", linestyle="-", linewidth=2,marker="*")

plt.title("My Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x) + 0.2 * np.random.randn(50)
```

```python
plt.plot(x,y1, label="sin(x)", color="green", linestyle="--", linewidth=2,marker="o")

plt.title("My Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend()
plt.show()
```
