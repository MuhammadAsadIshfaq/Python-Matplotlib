```python
import numpy as np
x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x) + 0.2 * np.random.randn(50)

plt.plot(x,y1, label="sin(x)", color="green", linestyle="--", linewidth=2,marker="o")

plt.title("My Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend()
plt.show()
```

```python
plt.scatter(x,y2, label="cos(x)", color="orange", linestyle="--", linewidth=2,marker="^",s=100)

plt.title("My Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

categories = np.array(['Category A', 'Category B', 'Category C', 'Category D'])
values = np.array([25, 40, 30, 35])

plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Vertical Bar Plot')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.barh(categories, values, color='lightcoral')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)

plt.figure(figsize=(8, 6))
plt.hist(data, bins=20, color='lightgreen', edgecolor='black', alpha=0.7)
plt.xlabel('Data Points')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [30, 25, 20, 15, 10]
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold', 'lightsalmon']
explode = (0.1, 0, 0, 0, 0) 

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=90)
plt.title('Proportions of Categories')
plt.show()
```
