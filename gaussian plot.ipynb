```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = np.array([2, 4, 6, 3, 8, 9, 11, 13, 14, 15, 18, 19, 20, 26, 28, 29, 31, 32, 35, 45, 41, 36, 33, 55, 44, 66, 77, 22, 34,
               54, 34, 34, 45, 32, 7, 34, 66, 43, 53, 12])

mean = np.mean(arr)
std = np.std(arr)
x = np.linspace(mean -3 *std, mean+3*std,100)

y = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-np.square(x - mean) / (2 * np.square(std)))
plt.plot(x, y, 'r--')
plt.show
```
